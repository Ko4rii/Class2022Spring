```python
r = response.json()
r
```

```python
items = r['response']['body']['tableList']['0']['items']['item']
items[0]
```

```python
title = [item['title'] for item in items]
crno = [item['crno'] for item in items]
fncoCd = [item['fncoCd'] for item in items]
fncoNm = [item['fncoNm'] for item in items]
xcsmCnt = [item['xcsmCnt'] for item in items]
xcsmDcd = [item['xcsmDcd'] for item in items]
xcsmDcdNm = [item['xcsmDcdNm'] for item in items]

title = pd.Series(title)
crno = pd.Series(crno)   
fncoCd = pd.Series(fncoCd)
fncoNm = pd.Series(fncoNm)
xcsmCnt = pd.Series(xcsmCnt)
xcsmDcd = pd.Series(xcsmDcd)
xcsmDcdNm = pd.Series(xcsmDcdNm)

df = pd.DataFrame({'title': title, 'crno': crno, 'fncoCd': fncoCd, 'fncoNm': fncoNm, 'xcsmCnt': xcsmCnt, 'xcsmDcd': xcsmDcd, 'xcsmDcdNm': xcsmDcdNm})
df
```
